```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset=load_breast_cancer()
```

```python
x=dataset.data
```

```python
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=123)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
```

Logistic Regresison is used for classification. Mostly used in Binary Classification. Here the target is either 0 or 1 that is malingnant or benign. So logistic regression can be a good model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
```

DecisionTreeClassifier is also used for classification purposes.This model can be a good fit because it uses flowchart like structures to make decisiond

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier()
```

Random forest Classifer is a good model because it avoids overfitting and also that it picks random feature

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3=RandomForestClassifier()
```

SVM is good for binary classification. With kernels, it can handle more complex data that isn’t linearly separable.

```python
from sklearn import svm
```

```python
model4=svm.SVC(C=1,gamma=0.7,kernel='rbf')
```

Makes no assumptions about the data’s shape, which is helpful for complex datasets.Works well with smaller datasets like this, where clear clusters might exist.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model5=KNeighborsClassifier(n_neighbors=3)
```

```python
classifier1=make_pipeline(StandardScaler(),model1)
classifier1.fit(x_train,y_train)
y_pred1=classifier1.predict(x_test)
ac1=accuracy_score(y_test,y_pred1)
cm1=confusion_matrix(y_test,y_pred1)
print("The accuracy score of Logistic regression model is",ac1,"\nThe confusion matrix is\n",cm1)
disp=ConfusionMatrixDisplay(cm1,display_labels=dataset.target_names)
disp.plot()
plt.show()
```

```python
classifier2=make_pipeline(StandardScaler(),model2)
classifier2.fit(x_train,y_train)
y_pred2=classifier2.predict(x_test)
ac2=accuracy_score(y_test,y_pred2)
cm2=confusion_matrix(y_test,y_pred2)
print("The accuracy score of Decision Tree Classifier model is",ac2,"\nThe confusion matrix is\n",cm2)
disp=ConfusionMatrixDisplay(cm2,display_labels=dataset.target_names)
disp.plot()
plt.show()
```

```python
classifier3=make_pipeline(StandardScaler(),model3)
classifier3.fit(x_train,y_train)
y_pred3=classifier3.predict(x_test)
ac3=accuracy_score(y_test,y_pred3)
cm3=confusion_matrix(y_test,y_pred3)
print("The accuracy score of Random Forest Classifier model is",ac3,"\nThe confusion matrix is\n",cm3)
disp=ConfusionMatrixDisplay(cm3,display_labels=dataset.target_names)
disp.plot()
plt.show()
```

```python
classifier4=make_pipeline(StandardScaler(),model4)
classifier4.fit(x_train,y_train)
y_pred4=classifier4.predict(x_test)
ac4=accuracy_score(y_test,y_pred4)
cm4=confusion_matrix(y_test,y_pred4)
print("The accuracy score of svm model is",ac4,"\nThe confusion matrix is\n",cm4)
disp=ConfusionMatrixDisplay(cm4,display_labels=dataset.target_names)
disp.plot()
plt.show()
```

```python
classifier5=make_pipeline(StandardScaler(),model5)
classifier5.fit(x_train,y_train)
y_pred5=classifier5.predict(x_test)
ac5=accuracy_score(y_test,y_pred5)
cm5=confusion_matrix(y_test,y_pred5)
print("The accuracy score of K Neighbors Classifier model is",ac5,"\nThe confusion matrix is\n",cm5)
disp=ConfusionMatrixDisplay(cm5,display_labels=dataset.target_names)
disp.plot()
plt.show()
```

From the accuracy scores we can see that Logistic regression, Random forest classifier and K neighbors classifier is a good model . All three models has accuracy score of 0.9912280701754386 

The worst algorithm/model is SVM , it has the accuracy score of 0.7105263157894737
